# While Loop

```python
n = int(input("Max iterations Your Number: "))
i = 1

while i < n:
    print(i)
    i += 1
    
print("Done")

```

```python
n = 10
i = 1
while True:
    if i%9 == 0:
        print("Inside if")
        break
    else:
        i +=1
        print("Inside else")
print("done")
```

```python
n = 10
i = 1
while True:
    if i%9 != 0:
        print("Inside if",i)
        i +=1
        continue
    print("the result",i)
    break
print("done")
```

# For Loop

```python
l= []
for i in range(10):
    print(i+1)
    l.append(i**2)
print(l)
```

```python
l= []
for i in range(1,20,3):
    print(i)
    l.append(i**2)
print(l)
```
